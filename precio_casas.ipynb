```python
#se divide la base datos
from sklearn.model_selection import train_test_split

X= data.drop(['median_house_value'], axis=1)
#print(X.head())
y=data['median_house_value']
print(y.head())
```

```python
#divisiÃ³n x train y x test, y train, y test
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
#Juntar base de datos de train
train_data=X_train.join(y_train)
```

```python
#normalizaciÃ³n del data train
train_data['total_rooms']= np.log(train_data['total_rooms']+1)
train_data['total_bedrooms']= np.log(train_data['total_bedrooms']+1)
train_data['population']= np.log(train_data['population']+1)
train_data['households']= np.log(train_data['households']+1)
```

```python
#cambiar variables categoricas a numerocas
train_data=train_data.join(pd.get_dummies(train_data['ocean_proximity']).astype(int)).drop(['ocean_proximity'], axis=1)
```

```python
#agregar variables
train_data['bedroom_ratio']=train_data['total_bedrooms']/train_data['total_rooms']
train_data['household_rooms']=train_data['total_rooms']/train_data['households']
```

```python
#mapa de calor
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True)
```

```python
#random forest para el train
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()

X_train,y_train= train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']
X_train_s=scaler.fit_transform(X_train)


reg= LinearRegression()

reg.fit(X_train_s, y_train)
```

```python
# crear mismos campos para el test set
test_data = X_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

test_data = test_data.join(pd.get_dummies(test_data['ocean_proximity']).astype(int)).drop(['ocean_proximity'], axis=1)

test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms'] / test_data['households']
```

```python
#igual campos para el test set
X_test,y_test= test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
#igualar campos para el test set
X_test_s=scaler.fit_transform(X_test)
```

```python
#random forest
from sklearn.ensemble import RandomForestRegressor

forest= RandomForestRegressor()

forest.fit(X_train_s, y_train)
```

```python
#r 2 del random forest
forest.score(X_test_s, y_test)
```

```python
#cross validation
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    "n_estimators": [100,200,300],
    "min_samples_split": [2,4],
    "max_features": [None, 4, 8]
}

grid_search = GridSearchCV(forest, param_grid, cv=5,
                          scoring="neg_mean_squared_error",
                          return_train_score=True)

grid_search.fit(X_train_s, y_train)
```

```python
grid_search.best_estimator_
```
